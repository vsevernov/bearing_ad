```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
data_dir = './data/row_data'

time_features = ['mean','std','rms']
cols = ['bearing_1', 'bearing_2', 'bearing_3', 'bearing_4']
columns = [c + '_' + tf for c in cols for tf in time_features]
data = pd.DataFrame(columns=columns)

# extract RMS features
def calculate_rms(df):
    result = []
    for col in df:
        r = np.sqrt((df[col]**2).sum() / len(df[col]))
        result.append(r)
    return result

# extract peak-to-peak features
def calculate_p2p(df):
    return np.array(df.max().abs() + df.min().abs())

merged_data = pd.DataFrame()
all_filename = os.listdir(data_dir)

if '.DS_Store' in all_filename:
    all_filename.remove('.DS_Store')
for filename in all_filename:
    dataset = pd.read_csv(os.path.join(data_dir, filename), sep='\t')
    dataset_mean = np.array(dataset.abs().mean())
    dataset_std = np.array(dataset.std())
    dataset_rms = np.array(calculate_rms(dataset))
    
    mean_df = pd.DataFrame(dataset_mean.reshape(1,4), columns=[col + '_mean' for col in cols])
    std_df = pd.DataFrame(dataset_std.reshape(1,4), columns=[col + '_std' for col in cols])
    rms_df = pd.DataFrame(dataset_rms.reshape(1,4), columns=[col + '_rms' for col in cols])

    mean_df.index = [filename]
    std_df.index = [filename]
    rms_df.index = [filename]
    
    merge = pd.concat([mean_df, std_df, rms_df], axis=1)
    data = data.append(merge)
```

```python
data.reset_index(inplace=True)
data = data.rename(columns = {'index':'time_key'})
```

```python
data.to_csv('bearing_features.csv', index=False)
```
