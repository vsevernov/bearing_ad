```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import matplotlib.dates as md
```

```python
DATA_PATH = "./bearing_features.csv"
```

```python
data = pd.read_csv(DATA_PATH)
data = data.set_index('time_key')
data.index = pd.to_datetime(data.index, format='%Y.%m.%d.%H.%M.%S')
data = data.sort_index()
```

```python
data.head()
```

```python
fig, ax = plt.subplots(figsize=(14, 6), dpi=80)
ax.plot(data['bearing_1_std'], label='bearing_1_std', color='blue', animated = True, linewidth=1)
ax.plot(data['bearing_2_std'], label='bearing_2_std', color='red', animated = True, linewidth=1)
ax.plot(data['bearing_3_std'], label='bearing_3_std', color='green', animated = True, linewidth=1)
ax.plot(data['bearing_4_std'], label='bearing_4_std', color='black', animated = True, linewidth=1)
plt.legend(loc='lower left')
ax.set_title('Bearing Sensor Training Data', fontsize=16)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14, 6), dpi=80)
ax.plot(data['bearing_1_mean'], label='bearing_1_mean', color='blue', animated = True, linewidth=1)
ax.plot(data['bearing_2_mean'], label='bearing_2_mean', color='red', animated = True, linewidth=1)
ax.plot(data['bearing_3_mean'], label='bearing_3_mean', color='green', animated = True, linewidth=1)
ax.plot(data['bearing_4_mean'], label='bearing_4_mean', color='black', animated = True, linewidth=1)
ax.xaxis.set_major_locator(md.HourLocator(interval = 3))
ax.xaxis.set_major_formatter(md.DateFormatter('%Y.%m.%d.%H'))
plt.legend(loc='lower left')
ax.set_title('Bearing Sensor Training Data', fontsize=16)
ax.tick_params(axis='x', labelrotation=45)
plt.show()
```
